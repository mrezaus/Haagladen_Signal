## Processing

```python
## Load section
df = pd.read_csv('Output/Org/Splitted_HRV_pts.csv')
```

```python
len(df)
```

```python
arrs = []
for item in set(df['Name']):
    arrs.append(item)
print(len(arrs))
arrs = sorted(list(arrs))
print(arrs)
```

```python
arr_n = []
for item in arrs:
    arr_n.append(len(df.loc[df['Name'] == item]))
print(arr_n)
```

```python
arr_z = []
for i in range(len(arr_n)):
    if i == 0:
        arr_z.append(arr_n[i])
    else:
        arr_z.append(arr_n[i]+arr_z[i-1])
print(arr_z)
print(len(arr_z))
```

```python
arr_cu = []
for i in range(len(arrs)):
    if i == 0:
        arr_cu.append([0,arr_z[i],arrs[i]])
    else:
        arr_cu.append([arr_z[i-1],arr_z[i],arrs[i]])
print(arr_cu)
```

```python
def Caller():
    for item in arr_cu:
        if i>=item[0] and i<item[1]:
            mi = i - item[1]
            lght = item[1] - item[0]
            ind = lght + mi
    return ind
```

```python
def loss(arr):
    tmp=0
    for i in range(len(arr)-1):
        tmp += (arr[i+1] - arr[i])**2
    m = len(arr)
    loss = 1/(2*m) * (tmp)
    return loss
```

```python
for i in range(len(df)):
    # for j in len(df.loc[df['Name'] == arrs[zx]]):

    text = df['Points'][i]
    text = ast.literal_eval(text)

    n1 = [text[z]/256 for z in range(0,len(text)-1)] #x
    n2 = [text[z]/256 for z in range(1,len(text))]   #y
    plt.scatter(n1,n2)
    fname = df['Name'][i]
    plt.title(f'{fname} - Number {Caller()+1}')
    plt.xlabel(df['Stage'][i])
    plt.savefig(f'Output/Scatter/{fname} - Number {Caller()+1}.png',dpi=100)
    plt.close()
    with open('Output/loss_splitted.csv','a') as file:
        file.write(f'{loss(text)}\n')
    # plt.show()
```
