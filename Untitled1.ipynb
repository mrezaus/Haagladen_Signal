```python
import mne
file = "SN001.edf"
data = mne.io.read_raw_edf(file)
raw_data = data.get_data()
# you can get the metadata included in the file and a list of all channels:
info = data.info
channels = data.ch_names

print(info)
print(channels)
print(type(raw_data))

# raw_data2 = mne.io.read_raw_edf("SN002.edf").get_data()
```

```python
wdata = raw_data[7][1000:3000]
# plt.yticks(np.arange(0.0002, 0.0008, step=0.0002))
plt.plot(wdata>0.0002)
print(np.count_nonzero(wdata > 0.0003))
```

```python
print(len(wdata[0:250]))
```

```python
ndata=[]
for i in wdata:
  if i > 0.0002:
    ndata.append(i)

countedhb=len(ndata)

plt.plot(ndata)
print(ndata)
```

```python
liz=[]
ip=0
for num in ndata[:(len(ndata)-1)]:
  if num < ndata[ip+1]:
    pass
  else:
    liz.append(num)
  ip += 1

liz
```

```python
plt.plot(ndata)
plt.plot(liz, 'r')
plt.show
```

```python
zer=[]
ip=0
for num in liz[:(len(liz)-1)]:
  if num < liz[ip+1]:
    pass
  else:
    zer.append(num)
  ip += 1

len(zer)
```

```python
plt.plot(ndata)
plt.plot(liz, 'r')
plt.plot(zer,'g')
plt.show
```

```python
ter=[]
ip=0
for num in zer[:(len(zer)-1)]:
  if num < zer[ip+1]:
    pass
  else:
    ter.append(num)
  ip += 1

len(ter)
```

```python
ler=[]
ip=0
for num in ter[:(len(ter)-1)]:
  if num < ter[ip+1]:
    pass
  else:
    ler.append(num)
  ip += 1

len(ler)
```

```python
plt.figure(dpi=400)

plt.plot(ndata)
plt.plot(liz, 'r')
plt.plot(zer,'g')
plt.plot(ter , 'brown')
plt.plot(ler, 'b')
plt.xlabel("Final Equation After 4 Iterations")
plt.savefig("Equation.png")

plt.show()
```
